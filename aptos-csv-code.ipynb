```python
import numpy as np
import cv2
import os
import csv
data_lebel=pd.read_csv('../input/aptos2019-blindness-detection/train.csv')


IMG_DIR = '../input/aptos2019-blindness-detection/train_images'

img_f=0

for img_name in os.listdir(IMG_DIR):
    print(img_name)
    print("")
    
    id_code=img_name.split('.')[0]
    
    print(id_code)
    
    label=data_lebel[data_lebel['id_code']==id_code]['diagnosis'].values[0]
    
    print(label)
    
    img_array = cv2.imread(os.path.join(IMG_DIR,img_name), cv2.IMREAD_GRAYSCALE)
    
    img_array=cv2.resize(img_array,(28,28))
    
    img_array=img_array.flatten()
    
    value=np.append(img_array,label)
    
    with open("aptos_dataset2.csv", 'a') as f:
        writer = csv.writer(f)
        writer.writerow(value)  
```

```python
import pandas as pd
data=pd.read_csv('../input/aptos-image-csv/aptos_dataset2.csv',header=None)
data

X=data.iloc[:,:-1]
y=data[784]
```

```python
from sklearn.cluster import KMeans
import numpy as np


# Initialize the K-Means model

kmeans = KMeans(n_clusters = 5)

# Fitting the model to training set

kmeans.fit(X)
```

```python
labels=kmeans.labels_
```

```python
y
```

```python
data['Cluster']=labels
```

```python
data['Cluster'].value_counts()
```

```python
data=data.rename({784: 'Actual'}, axis=1)  # new method
```

```python
data
```

```python
from sklearn import metrics

metrics.accuracy_score(data.Actual, data.Cluster)
```
